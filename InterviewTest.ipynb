```python
import os
import glob
import pandas

def concatenate(indir="G:\\intership-test-master\\output\\filterData\\",outfile="G:\\intership-test-master\\output\\lowestPrice.csv"):
    os.chdir(indir)
    fileList=glob.glob("data_*.csv")
    dfList=[]
    colnames=["SKU","FIRST_MINIMUM_PRICE","SECOND_MINIMUM_PRICE"]
    for filename in fileList:
        print(filename)
        df=pandas.read_csv(filename,header=None)
        dfList.append(df)
    concatDf=pandas.concat(dfList,axis=1)
    concatDf.columns=colnames
    concatDf.to_csv(outfile,index=None)    
    
concatenate()
```
